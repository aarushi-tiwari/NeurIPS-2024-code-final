```python
import tensorflow as tf

from __future__ import absolute_import, division, print_function, unicode_literals
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop, Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
#import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.regularizers import l2
```

```python
#load output feature maps
blank = np.load('/content/blank_unbiased_2.npy 2')
target = np.load('/content/target_unbiased_2.npy 2')


print(blank.shape)
print(blank.shape)
blank_train = blank[0:3000]
blank_val = blank[3000:4500]



target_train = target[0:3000]
target_val = target[3000:4500]

blank_test = blank[4500:]
target_test = target[4500:]
```

```python
from tensorflow import keras
from keras.regularizers import l2

# define model
model = keras.models.Sequential([
    keras.layers.Conv2D(10, (10,10), activation='relu', input_shape=(34,34,1)),
    keras.layers.MaxPooling2D(3,3),
    keras.layers.Flatten(),
    keras.layers.Dense(220, activation='relu'),
    keras.layers.Dense(230, activation='relu'),
    keras.layers.Dense(240, activation='relu'),
    keras.layers.Dense(200, activation='relu'),
    keras.layers.Dense(180, activation='relu'),
    keras.layers.Dense(160, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
             optimizer=RMSprop(learning_rate= 0.0005, momentum = 0.9),
             metrics=['acc'])
```

```python
batch_size = 450
num_classes = 1
epochs = 400
history = model.fit(blank_train, target_train.reshape(3000,1),
              batch_size=batch_size,
              epochs=epochs,
              verbose=1,
              validation_data=(blank_val, target_val))
```

```python

test_loss, test_accuracy = model.evaluate(blank_test, target_test)
print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```
