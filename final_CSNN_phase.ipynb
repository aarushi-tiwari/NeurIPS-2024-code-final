```python


def reshape_spectrogram_to_square(spectrogram, target_side_length):

     rows, cols = spectrogram.shape

     pad_rows = max(0, target_side_length - rows)
     pad_cols = max(0, target_side_length - cols)


     padded_spectrogram = np.pad(spectrogram, ((round(pad_rows/2), pad_rows - round(pad_rows/2)), (round(pad_cols/2), pad_cols - round(pad_cols/2))), mode='constant', constant_values=0)
     cropped_spectrogram = np.flip(padded_spectrogram[:target_side_length, :target_side_length], axis = 0)

     return cropped_spectrogram
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import SpykeTorch.utils as utils
import SpykeTorch.functional as sf
import torchvision.transforms as transforms
import numpy as np
from PIL import Image

# Create data_loader object to apply spike wave transform to images

class MyDataset(Dataset):
    def __init__(self, data, target, transform):
        self.data = data
        self.target = target
        self.transform = transform

    def __getitem__(self, index):
        x = self.data[index].reshape(150,150)
        y = self.target[index]

        if self.transform:
            x = self.transform(Image.fromarray(x))

        return x, y

    def __len__(self):
        return len(self.data)

    def myTransform(self, index):
        x = self.data[index].reshape(150,150)
        y = self.target[index]


        if self.transform:
           self.transform(Image.fromarray(x))

        return x, y


def time_dim(input):
    return input.unsqueeze(0)

import SpykeTorch.utils as utils

kernels = [	utils.GaborKernel(window_size = 3, orientation = 45+22.5),
            utils.GaborKernel(3, 90+22.5),
            utils.GaborKernel(3, 135+22.5),
            utils.GaborKernel(3, 180+22.5)]
filter = utils.Filter(kernels, use_abs = True)
```

```python
dataset_train = MyDataset(X[train_indices], y[train_indices], transform = transforms.Compose(
    [transforms.Grayscale(),
    transforms.ToTensor(),
    time_dim,
    filter,
    sf.pointwise_inhibition,
    utils.Intensity2Latency(number_of_spike_bins = 15, to_spike = True)]))


dataset_test = MyDataset(X[test_indices], y[test_indices], transform = transforms.Compose(
    [transforms.Grayscale(),
    transforms.ToTensor(),
    time_dim,
    filter,
    sf.pointwise_inhibition,
    utils.Intensity2Latency(number_of_spike_bins = 15, to_spike = True)]))
```

```python
from torch.utils.data import DataLoader
from torch.utils.data import SubsetRandomSampler

dataset_train = utils.CacheDataset(dataset_train)
dataset_test = utils.CacheDataset(dataset_test)

train_loader = DataLoader(dataset_train,  batch_size = 3, shuffle = False)
test_loader = DataLoader(dataset_test, shuffle = False)

```

```python
import SpykeTorch.snn as snn

pool = snn.Pooling(kernel_size = 3, stride = 2)
```

```python
#Not sure what to make kernel size
conv = snn.Convolution(in_channels=4, out_channels= 54, kernel_size=45)
```

```python
conv.reset_weight()
anti_stdp = snn.STDP(conv_layer = conv, learning_rate = (-0.05, 0.0005))
```

```python
stdp = snn.STDP(conv_layer = conv, learning_rate = (0.05, -0.015))
```

```python
feature2class = [0] * 27 + [1] * 27
print(feature2class)
```

```python
# reinforcement learning of CSNN
print("Starting Reinforcement Learning ...")
for iter in range(3):
    print('\rIteration:', iter, end="")
    for data,targets in train_loader:
        #print(targets)

        for x,t in zip(data, targets):
            x = sf.pad(x, (5,5,5,5))
            x = pool(x)
            p = conv(x)

            o, p = sf.fire(p, 20, return_thresholded_potentials=True)
            print(o.shape)

            winners = sf.get_k_winners(p, kwta=1, inhibition_radius=0, spikes=o)
            if len(winners) != 0:
                if feature2class[winners[0][0]] == t:
                    stdp(x, p, o, winners)
                else:
                    anti_stdp(x, p, o, winners)



print("Reinforcement Learning is Done.")
```

```python

error = 0
silent = 0
total = 0
total2 = 0
# testing CSNN and collecting output feature maps
blank = np.zeros((6000, 34, 34))
target_blank = np.zeros(6000)
x_save = np.zeros((6000,34, 34))
for data,targets in test_loader:
    for x,t in zip(data, targets):
        total += 1

        x = sf.pad(x, (5,5,5,5))

        x = pool(x)

        p2 = conv(x)

        result = torch.sum(p2, dim=0)
        result = result[0]

        print(result.shape)

        o, p = sf.fire(p2, 20, return_thresholded_potentials=True)
        total2+=1
        blank[total2-1, :, :] = result
        target_blank[total2-1] = t
        print(blank.shape)


        winners = sf.get_k_winners(p, kwta=1, inhibition_radius=0, spikes=o)
        if len(winners) != 0:
            if feature2class[winners[0][0]] != t:
                error += 1
        else:
            silent += 1
print("         Error:", error/total)
print("Silent Samples:", silent/total)
```
